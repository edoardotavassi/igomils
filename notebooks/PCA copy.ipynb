```python
#delete last two columns
df = df.drop(['CCI'], axis=1)
```

```python
class_cci = df['Class_CCI']
df = df.drop(['Class_CCI'], axis=1)
```

```python
class_cci.unique()
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df = df_scaled
```

```python
#apply pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)

#transform the data
df_pca = pca.transform(df)

#plot the data with respect to the 5 cci classes
plt.figure(figsize=(10, 10))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=class_cci, cmap='viridis')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.title('Data after PCA')
plt.colorbar()
plt.show()




```

```python
#apply t-sne
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(df)

#plot the data in two separate graph for each class
fig, ax = plt.subplots(1, 3, figsize=(15, 6))

for i in range(3):

    ax[i].scatter(df_tsne[class_cci == i, 0], df_tsne[class_cci == i, 1], c='blue', label='Class 0')
    ax[i].scatter(df_tsne[class_cci == i+1, 0], df_tsne[class_cci == i+1, 1], c='red', label='Class 1')
    ax[i].scatter(df_tsne[class_cci == i+2, 0], df_tsne[class_cci == i+2, 1], c='green', label='Class 2')
    ax[i].scatter(df_tsne[class_cci == i+3, 0], df_tsne[class_cci == i+3, 1], c='yellow', label='Class 3')
    ax[i].scatter(df_tsne[class_cci == i+4, 0], df_tsne[class_cci == i+4, 1], c='purple', label='Class 4')
    ax[i].set_title(f'Class {i}')
    ax[i].legend()

plt.show()
```

```python
#apply umap
import umap
import umap.plot
umap_model = umap.UMAP(metric='euclidean', random_state=42, low_memory=True)
mapper = umap_model.fit(df)

# Plot the data using the fitted UMAP model with `conversione` as the color label
plt.figure(figsize=(10, 8))
umap.plot.points(mapper, labels=class_cci, theme='viridis')

plt.title('UMAP Projection with Conversione Labels')
plt.show()
```

```python
import featuretools as ft
import pandas as pd
from featuretools.primitives import Mean, Sum, MultiplyNumeric


df['Class_CCI'] = class_cci

es = ft.EntitySet(id="medical_data")
df["index"] = df.index
es = es.add_dataframe(dataframe_name="patients", dataframe=df, index="index")

# Specify the primitives to use
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="patients",
    agg_primitives=[Mean, Sum],
    trans_primitives=[MultiplyNumeric],
    ignore_columns={"patients": ["Class_cci"]}  # Ensure conversione is not used as a feature
)

feature_matrix.head()
```
