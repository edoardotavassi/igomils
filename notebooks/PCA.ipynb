```python
#delete last two columns
df = df.drop(['CCI', 'Class_CCI'], axis=1)
```

```python
#detach Conversione column
conversione = df['Conversione']
df = df.drop(['Conversione'], axis=1)
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df = df_scaled
```

```python
#apply pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)

#transform the data
df_pca = pca.transform(df)

#plot the data in two separate graph for each class
fig, ax = plt.subplots(1, 3, figsize=(15, 6))

ax[0].scatter(df_pca[conversione == 0, 0], df_pca[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[0].scatter(df_pca[conversione == 1, 0], df_pca[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[0].set_title('Conversione')
ax[0].legend()


ax[1].scatter(df_pca[conversione == 0, 0], df_pca[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[1].set_title('No Conversione')
ax[1].legend()

ax[2].scatter(df_pca[conversione == 1, 0], df_pca[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[2].set_title('Conversione')
ax[2].legend()

plt.show()

```

```python
#apply t-sne
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(df)

#plot the data in two separate graph for each class
fig, ax = plt.subplots(1, 3, figsize=(15, 6))

ax[0].scatter(df_tsne[conversione == 0, 0], df_tsne[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[0].scatter(df_tsne[conversione == 1, 0], df_tsne[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[0].set_title('Conversione')
ax[0].legend()


ax[1].scatter(df_tsne[conversione == 0, 0], df_tsne[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[1].set_title('No Conversione')
ax[1].legend()

ax[2].scatter(df_tsne[conversione == 1, 0], df_tsne[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[2].set_title('Conversione')
ax[2].legend()

plt.show()
```

```python
#apply umap
import umap
import umap.plot
umap_model = umap.UMAP(metric='euclidean', random_state=42, low_memory=True)
mapper = umap_model.fit(df)

# Plot the data using the fitted UMAP model with `conversione` as the color label
plt.figure(figsize=(10, 8))
umap.plot.points(mapper, labels=conversione, theme='viridis')

plt.title('UMAP Projection with Conversione Labels')
plt.show()
```

```python
import featuretools as ft
import pandas as pd
from featuretools.primitives import Mean, Sum, MultiplyNumeric


df['conversione'] = conversione

es = ft.EntitySet(id="medical_data")
df["index"] = df.index
es = es.add_dataframe(dataframe_name="patients", dataframe=df, index="index")

# Specify the primitives to use
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="patients",
    agg_primitives=[Mean, Sum],
    trans_primitives=[MultiplyNumeric],
    ignore_columns={"patients": ["conversione"]}  # Ensure conversione is not used as a feature
)

feature_matrix.head()
```

```python
umap_model = umap.UMAP(metric='euclidean', random_state=42, low_memory=True)
mapper = umap_model.fit(feature_matrix)

# Plot the data using the fitted UMAP model with `conversione` as the color label
plt.figure(figsize=(10, 8))
umap.plot.points(mapper, labels=conversione, theme='viridis')

plt.title('UMAP Projection with Conversione Labels')
plt.show()
```

```python
#apply t-sne
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(feature_matrix)

#plot the data in two separate graph for each class
fig, ax = plt.subplots(1, 3, figsize=(15, 6))

ax[0].scatter(df_tsne[conversione == 0, 0], df_tsne[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[0].scatter(df_tsne[conversione == 1, 0], df_tsne[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[0].set_title('Conversione')
ax[0].legend()


ax[1].scatter(df_tsne[conversione == 0, 0], df_tsne[conversione == 0, 1], label='No Conversione', c='red', alpha=0.5)
ax[1].set_title('No Conversione')
ax[1].legend()

ax[2].scatter(df_tsne[conversione == 1, 0], df_tsne[conversione == 1, 1], label='Conversione', c='blue', alpha=0.5)
ax[2].set_title('Conversione')
ax[2].legend()

plt.show()
```
