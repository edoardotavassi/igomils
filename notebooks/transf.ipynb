```python
import pandas as pd
df = pd.read_csv('../data/processed/igomilsdb_s_cleaned.csv')
df.head()
```

```python
#drop class_CCI column and CCI column
df = df.drop(['Class_CCI', 'CCI'], axis=1)
df.head()
```

```python
conversione = df['Conversione']
df = df.drop(['Conversione'], axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
#train test split
X_train, X_test, y_train, y_test = train_test_split(df, conversione, test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTENC

categorical_features = [0, 3, 4, 5, 6, 7, 8, 12, 13, 15, 16]


#rebalance the training set
sm = SMOTENC(random_state=42, categorical_features= categorical_features)
X_res, y_res = sm.fit_resample(X_train, y_train)

#lower the number of synthetic samples

y_res.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

encoder = OneHotEncoder(sparse_output=False)
scaler = StandardScaler()

# List of categorical columns to encode
categorical_features = [0, 3, 4, 5, 6, 7, 8, 12, 13, 15, 16]


# Identify continuous columns by excluding categorical columns
continuous_features = [col for col in df.columns if col not in X_res.columns[categorical_features]]



# One-hot encode the categorical columns
encoded = encoder.fit_transform(X_res.iloc[:, categorical_features])

# Convert encoded array to a DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(X_res.columns[categorical_features]))

# Extract and scale the continuous columns
continuous_df = pd.DataFrame(scaler.fit_transform(X_res[continuous_features]), columns=continuous_features)

# Concatenate the scaled continuous columns, one-hot encoded DataFrame, and the dropped column
X_train = pd.concat([continuous_df, encoded_df], axis=1)

y_train = y_res

# Display the first few rows of the final DataFrame
X_train.head()
```

```python
# One-hot encode the categorical columns
encoded = encoder.fit_transform(X_test.iloc[:, categorical_features])

# Convert encoded array to a DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(X_test.columns[categorical_features]))

# Extract and scale the continuous columns
continuous_df = pd.DataFrame(scaler.fit_transform(X_test[continuous_features]), columns=continuous_features)

# Concatenate the scaled continuous columns, one-hot encoded DataFrame, and the dropped column
X_test = pd.concat([continuous_df, encoded_df], axis=1)

# Display the first few rows of the final DataFrame
X_test.head()
```

```python
missing_cols = set(X_train.columns) - set(X_test.columns)
#add the missing columns to the test set
for col in missing_cols:
    X_test[col] = 0
```

```python
import torch
# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)
```

```python
import torch.nn as nn
import torch.optim as optim

import torch.nn as nn

class SimpleTransformer(nn.Module):
    def __init__(self, input_dim, projection_dim, num_heads, num_layers, hidden_dim):
        super(SimpleTransformer, self).__init__()
        
        # Linear layer to project input_dim to projection_dim
        self.input_projection = nn.Linear(input_dim, projection_dim)
        
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=projection_dim,  # Use projection_dim as the d_model
            nhead=num_heads, 
            dim_feedforward=hidden_dim
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # Linear layer to output a single value after pooling
        self.fc = nn.Linear(projection_dim, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        x = self.input_projection(x)  # Project the input to the desired dimension
        x = self.transformer_encoder(x.unsqueeze(1))  # Apply transformer
        x = x.mean(dim=0)  # Mean pooling over the sequence dimension (reduce to [batch_size, d_model])
        x = self.fc(x)  # Apply linear layer
        return self.sigmoid(x).squeeze(1)  # Match the output size to the number of samples

# Instantiate the model
model = SimpleTransformer(
    input_dim=X_train_tensor.shape[1],  # Number of features (should be 20 in this case)
    projection_dim=64,  # Desired dimension to match d_model
    num_heads=4,    # Number of heads
    num_layers=2,   # Number of transformer layers
    hidden_dim=64   # Hidden layer size
)
```

```python
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 10
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    
    # Forward pass
    outputs = model(X_train_tensor)
    
    # Calculate loss
    loss = criterion(outputs, y_train_tensor)
    
    # Backward pass
    loss.backward()
    optimizer.step()
    
    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')

    # Evaluation (on test set)
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test_tensor)
        test_loss = criterion(test_outputs, y_test_tensor)
        print(f'Test Loss: {test_loss.item()}')
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Convert outputs to binary predictions
preds = (test_outputs > 0.5).float()

# Convert predictions and labels to numpy arrays for evaluation
y_pred = preds.numpy()
y_true = y_test_tensor.numpy()

print(classification_report(y_true, y_pred))
print(confusion_matrix(y_true, y_pred))
```
